```
face_classifier=cv2.CascadeClassifier('/content/drive/My Drive/docs/scan/AI/haarcascade_eye.xml')
```

```
image=cv2.imread('/content/drive/My Drive/docs/scan/AI/barack-obama.jpeg')
```

```
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```
gray
```

```
faces=face_classifier.detectMultiScale(gray,1.3,6)
```

```
faces
```

```
if faces is ():
  print("no face detect")
```

```
for (x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w,y+h),(125,152,255),5)
  cv2_imshow(image)
  cv2.waitKey(0)
 
```
